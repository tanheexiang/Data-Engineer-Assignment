```python
import pandas as pd
import numpy as np

file_name = 'C:/Users/b2tanhe/Desktop/Sales_Transactions_Dataset_Weekly.csv'

df = pd.read_csv(file_name)
df.head()
```

```python
#Only select column 'Product_Code' and the 52 weeks columns for further processing

new_df = df.iloc[:,1:53]
print(new_df)
```

```python
#Identified the 99% quartile value of each products

std_dev_df = new_df.quantile(q=0.99, axis=1)

print(std_dev_df)
```

```python
#Added in the existing product_code/weeks dataframe with the 99% quartile result

new_df['std_dev'] = std_dev_df

print(new_df)
```

```python
#For each product, compare all the week sales volume to the 99% quartile value.
#If there is one week sales value larger than the 99% quartile value, record the week number.
#Else, that product has no outlier sales on a particular week.

outlier_week = std_dev_df

for i in range (0,811):
    outlier_week[i] = ''

for i in range(0,52):
    for j in range (0, 811):
        if new_df.iloc[j]['W'+str(i)] > new_df.iloc[j]['std_dev']:
            outlier_week[j] = 'W'+str(i)

print('The outlier data corresponding to its week of each product is as listed in ascending order:')
print(outlier_week)
```
