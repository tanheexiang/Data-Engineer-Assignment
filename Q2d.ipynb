```python
import pandas as pd
import numpy as np
file_name = 'C:/Users/b2tanhe/Desktop/Sales_Transactions_Dataset_Weekly.csv'

df = pd.read_csv(file_name)
df.head()
```

```python
#Only select column 'Product_Code' and the 52 weeks columns for further processing

new_df = df.iloc[:,:53]
print(new_df)
```

```python
#Biweekly basis performance is defined as every 2 weeks as a pair in sales.
#So, for product P1, for values of W1, the sales volume is added in one week prior and one week after it.
#For example, product P1, W1 has 12 sales, in biweekly basis performance, W0+W1(Prior pair) volume is 11+12=23; W1+W2(Posterior) volume is 12+10=22
#Do this for all the weeks and for all the products with this function.

def Biweekly_Sum (df, i):
    
    val = 0
    
    columns = ['W'+str(i) for i in range(i, i+2)]
    
    print(i+2)
    
    total_df = df[columns].sum(axis=1)
    
    return total_df
```

```python
#The result of the biweekly basis performance as defined above with example of the last week - week 51
#Since after W51 there is no W52, the only biweekly pair related to W51 for P1 product sales volume is W50+W51=10+5=15(Prior)
#The result is the sum of all the sales volumes of all the products with biweekly pair W51.

print (Biweekly_Sum(new_df, 50))
```

```python
#Combine the dataframe with biweekly sales volumes

i = 0 

BiWeekly = 'Bi'

for i in range (0, 51):
    new_df[BiWeekly + str(i)] = Biweekly_Sum(new_df, i)

bi_df = new_df
print(bi_df)
```

```python
#Select only biweekly columns, without the product code and original sales volume data per week
bi_df = new_df.iloc[: ,54:104]
print(bi_df)
```

```python
#Sum all the biweekly sales volumn for each row

columns = ['Bi'+str(i) for i in range(1, 51)]

total_bi_df = bi_df[columns].sum(axis=1)

print(total_bi_df)
```

```python
#Identify the total minimal sum value

total_bi_df.min()
```

```python
#Identify the index position of the total minimal sum value

total_bi_df.idxmin(axis=0)
```

```python
#Locate the product with the index position value

biweekly_worst_product = df.iloc[[total_bi_df.idxmin(axis=0)]]
print('Worst performing product on a biweekly basis is', biweekly_worst_product['Product_Code'])
```
