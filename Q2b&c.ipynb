```python
import pandas as pd
import numpy as np
file_name = 'C:/Users/b2tanhe/Desktop/Sales_Transactions_Dataset_Weekly.csv'

df = pd.read_csv(file_name)
df.head()
```

```python
#Only select column 'Product_Code' and the 52 weeks columns for further processing

new_df = df.iloc[:,:53]
print(new_df)
```

```python
#Sum all the number of sales for 52 weeks of all products

columns = ['W'+str(i) for i in range(0, 52)]

total_df = new_df[columns].sum(axis=1)

print(total_df)
```

```python
# Get the max value of all the total sales of all products

total_df.max()
```

```python
#Identify the index position of the highest sales product

total_df.idxmax(axis=0)
```

```python
#Locate the product based on the index position

best_product = df.iloc[[total_df.idxmax(axis=0)]] 

print('Best performing product is', best_product['Product_Code'])
```

```python
#Question 2c
#Define promising products as the top 5% highest selling products
#Start by sorting the total sum of all the products

sorted_df = total_df.sort_values(axis=0, ascending=False)

print(sorted_df)
```

```python
pd.qcut(sorted_df, [0.95])
print(pd.qcut(sorted_df, [0.95]))
```
